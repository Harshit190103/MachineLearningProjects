# Data Preprocessing

```python
X=ds.iloc[:,:-1]
y= ds.iloc[:,-1]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

# Feature Rescaling

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
boston.feature_names
```

```python
# the respective coeffecints wrt to features
print(model.coef_)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
plt.scatter(y_test,pred,edgecolors='orange')
```

```python
# lets check the residuals

residuals=pred-y_test
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
## scatter plot with respect to predication and residuals
plt.scatter(pred,residuals)
```

some of the points are here and there but most of the points are uniformly distributed

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(pred,y_test))
print(mean_squared_error(pred,y_test))
```

# R Square And Adjusted R Square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)
print(score)
```

```python
adj_r2=1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
print(adj_r2)
```
